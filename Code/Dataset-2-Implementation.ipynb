```python
import numpy as np
import pandas as pd
from sklearn.experimental import enable_iterative_imputer  # Enable IterativeImputer
from sklearn.impute import SimpleImputer, IterativeImputer
from sklearn.metrics import mean_squared_error, mean_absolute_error, davies_bouldin_score
from sklearn.preprocessing import PowerTransformer
from sklearn.covariance import LedoitWolf
from scipy.cluster.hierarchy import linkage, fcluster
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.covariance import OAS

# Load Dataset
file_path = '/kaggle/input/datasetnum2/heart.csv'
data = pd.read_csv(file_path)
print("Dataset Loaded Successfully!")

# Prepare numerical features for preprocessing
numerical_features = data.select_dtypes(include=[np.number]).columns
processed_features = numerical_features.tolist()
original_data = data.copy()

# Apply PowerTransformer (or choose another scaler if necessary)
scaler = PowerTransformer(method='yeo-johnson', standardize=True)
scaled_data = data.copy()
scaled_data[processed_features] = scaler.fit_transform(data[processed_features])

# Function to introduce missing values
def introduce_missingness(data, features, missingness):
    data_with_missing = data.copy()
    num_samples, num_features = data_with_missing[features].shape
    missing_count = int((missingness / 100) * num_samples * len(features))
    missing_indices = [
        (np.random.randint(0, num_samples), np.random.choice(range(len(features))))
        for _ in range(missing_count)
    ]
    for row, col in missing_indices:
        feature = features[col]
        data_with_missing.at[row, feature] = np.nan
    return data_with_missing

# Mean Imputation
def mean_imputation(data, features):
    imputer = SimpleImputer(strategy='mean')
    imputed_data = imputer.fit_transform(data[features])
    return pd.DataFrame(imputed_data, columns=features)

# Hot Deck Imputation (using nearest available value)
def hot_deck_imputation(data, features):
    imputed_data = data.copy()
    for feature in features:
        imputed_data[feature].fillna(method='ffill', inplace=True)
        imputed_data[feature].fillna(method='bfill', inplace=True)
    return imputed_data[features]

# MICE Imputation
def mice_imputation(data, features):
    mice_imputer = IterativeImputer(random_state=42, max_iter=20, tol=1e-3)
    imputed_data = mice_imputer.fit_transform(data[features])
    return pd.DataFrame(imputed_data, columns=features)

def devised_imputation(data, features, missingness):
    # Step 1: Compute the initial covariance matrix from the original data
    initial_covariance = np.cov(data[features], rowvar=False)

    # Step 2: Introduce missingness into the dataset
    data_with_missing = introduce_missingness(data, features, missingness)
    scaled_missing_data = data_with_missing[features]

    # Step 3: Perform initial imputation using Iterative Imputer with GradientBoostingRegressor
    advanced_iterative_imputer = IterativeImputer(
        estimator=GradientBoostingRegressor(n_estimators=100, random_state=42),
        max_iter=20,
        random_state=42,
        imputation_order="ascending",
        tol=1e-3
    )
    imputed_data = advanced_iterative_imputer.fit_transform(scaled_missing_data)

    # Step 4: Perform hierarchical clustering on the initially imputed data
    linkage_matrix = linkage(imputed_data, method='ward')
    clusters = fcluster(linkage_matrix, t=4, criterion='maxclust')

    # Step 5: Adjust missing values in each cluster using covariance adjustments
    missing_mask = data_with_missing[features].isna()
    adjusted_data = imputed_data.copy()

    for cluster in np.unique(clusters):
        # Identify rows belonging to the current cluster
        cluster_mask = clusters == cluster
        cluster_data = imputed_data[cluster_mask]

        # Compute the cluster-specific covariance matrix
        lw = LedoitWolf()
        cluster_covariance = lw.fit(cluster_data).covariance_

        # Compute the covariance difference
        covariance_difference = initial_covariance - cluster_covariance

        # Adjust missing values for each feature in this cluster
        for feature_idx, feature in enumerate(features):
            missing_rows = missing_mask[feature] & cluster_mask
            if missing_rows.any():
                # Calculate the adjustment factor based on covariance difference
                adjustment_factor = covariance_difference[feature_idx, feature_idx]

                # Adjust the imputed values to align with original covariance
                cluster_mean = cluster_data[:, feature_idx].mean()
                adjusted_values = (
                    adjusted_data[missing_rows, feature_idx] 
                    + adjustment_factor 
                    * (adjusted_data[missing_rows, feature_idx] - cluster_mean)
                )
                adjusted_data[missing_rows, feature_idx] = adjusted_values

    # Return the final adjusted dataset as a DataFrame
    return pd.DataFrame(adjusted_data, columns=features)

# Define methods and results storage
methods = {
    "Mean": mean_imputation,
    "Hot Deck": hot_deck_imputation,
    "MICE": mice_imputation,
    "Proposed": devised_imputation
}
missingness_levels = [10, 20, 30, 40, 50]
results_mae = {method: [] for method in methods}
results_mse = {method: [] for method in methods}
results_rmse = {method: [] for method in methods}
dbi_results = {method: [] for method in methods}

# Run imputation methods for different missingness levels
for missingness in missingness_levels:
    print(f"\nProcessing {missingness}% Missingness...")
    data_with_missing = introduce_missingness(scaled_data, processed_features, missingness)

    for method_name, imputation_function in methods.items():
        if method_name == "Proposed":
            imputed_data = imputation_function(scaled_data, processed_features, missingness)
        else:
            imputed_data = imputation_function(data_with_missing, processed_features)

        # Calculate MSE and MAE
        mse = mean_squared_error(scaled_data[processed_features], imputed_data)
        mae = mean_absolute_error(scaled_data[processed_features], imputed_data)
        rmse = np.sqrt(mse)
        results_mae[method_name].append(mae)
        results_mse[method_name].append(mse)
        results_rmse[method_name].append(rmse)

        # Perform hierarchical clustering
        linkage_matrix = linkage(imputed_data, method='ward')
        clusters = fcluster(linkage_matrix, t=4, criterion='maxclust')

        # Calculate DBI
        dbi = davies_bouldin_score(imputed_data, clusters)
        dbi_results[method_name].append(dbi)

# Print results
print("\nResults (MAE):")
for method, mae_values in results_mae.items():
    print(f"{method}: {mae_values}")

print("\nResults (MSE):")
for method, mse_values in results_mse.items():
    print(f"{method}: {mse_values}")

print("\nResults (RMSE):")
for method, rmse_values in results_rmse.items():
    print(f"{method}: {rmse_values}")

print("\nResults (DBI):")
for method, dbi_values in dbi_results.items():
    print(f"{method}: {dbi_values}")

# Define line styles or markers for each method
line_styles = {
    "Mean": "o-",         # Circle marker with solid line
    "Hot Deck": "s-",    # Square marker with solid line
    "MICE": "d-",        # Diamond marker with solid line
    "Proposed": "^-",     # Triangle marker with solid line
}

# Plot MAE
plt.figure(figsize=(8, 6))
for method, mae_values in results_mae.items():
    plt.plot(
        missingness_levels,
        mae_values,
        line_styles[method],
        label=f"{method}",
        markersize=6
    )
plt.title("MAE vs Missingness Level")
plt.xlabel("Missingness Level (%)")
plt.ylabel("Mean Absolute Error")
plt.xticks(range(10, 55, 5))  # Set x-axis ticks
plt.legend()
plt.show()

# Plot MSE
plt.figure(figsize=(8, 6))
for method, mse_values in results_mse.items():
    plt.plot(
        missingness_levels,
        mse_values,
        line_styles[method],
        label=f"{method}",
        markersize=6
    )
plt.title("MSE vs Missingness Level")
plt.xlabel("Missingness Level (%)")
plt.ylabel("Mean Squared Error")
plt.xticks(range(10, 55, 5))  # Set x-axis ticks
plt.legend()
plt.show()

# Plot RMSE
plt.figure(figsize=(8, 6))
for method, rmse_values in results_rmse.items():
    plt.plot(
        missingness_levels,
        rmse_values,
        line_styles[method],
        label=f"{method}",
        markersize=6
    )
plt.title("RMSE vs Missingness Level")
plt.xlabel("Missingness Level (%)")
plt.ylabel("Root Mean Squared Error")
plt.xticks(range(10, 55, 5))  # Set x-axis ticks
plt.legend()
plt.show()
```
